```python
from gensim.summarization import summarize
from gensim.summarization import keywords
import requests
import logging 
import re
import matplotlib.pyplot as plt
import numpy as np
import spacy

logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO) 
```

```python
from spacy.en import English
nlp = spacy.load("en")
```

```python
my_stop_words = [u'say', u'\'s', u'Mr', u'be', u'said', u'says', u'saying']
for stopword in my_stop_words:
    lexeme = nlp.vocab[stopword]
    lexeme.is_stop = True
```

```python
r = requests.get('https://www.archives.gov/founding-docs/bill-of-rights-transcript')
from bs4 import BeautifulSoup  
soup = BeautifulSoup(r.text, 'lxml')


for result in soup.find_all(class_="region region-content"):
    text = result.get_text('\n')
    print(text)
       
```

```python
print(len(text))
```

```python
doc = nlp(text)
```

```python
print ('Summary:')
print (summarize(text))
```

```python
#Use the "split" option if you want a list of strings instead of a single string.
print (summarize(text, split=True))
```

```python
# Adjust how much text the summarizer outputs via the "ratio" parameter or the "word_count" parameter. 
# Using the "ratio" parameter, you specify what fraction of sentences in the original text should be returned as output. 
print ('Summary:')
print (summarize(text, ratio=0.2))
```

```python
# Specify the maximum amount of words we want in the summary. 

print ('Summary:')
print (summarize(text, word_count=200))
```

```python
#Extract the keywords
print ('\nKeywords:')
print (keywords(text))
```
