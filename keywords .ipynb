```python
import spacy
from spacy.en import English
nlp = spacy.load("en")
from collections import Counter, defaultdict
import requests
import logging 
```

```python
r = requests.get('https://www.archives.gov/founding-docs/stylistic-artistry-of-the-declaration')
from bs4 import BeautifulSoup  
soup = BeautifulSoup(r.text, 'lxml')


for result in soup.find_all(class_="region region-content"):
    text = result.get_text('\n')
      
```

```python
text
```

```python
doc = nlp(text.strip())
```

```python
doc
```

```python
# Get first token of the processed document
token = doc[0]
print('Token: ', token)
# Print all the tokens
tokens = [token.orth_ for token in doc if not token.orth_.isspace()]
print("There were {} tokens found. Here's a sample:".format(len(tokens)))
# Print sentences (one sentence per line)
for sent in doc.sents:
    print('Sentences:')
    print(sent)
```

```python
sentences = [sentence.orth_ for sentence in doc.sents if not sentence.orth_.isspace()]
print("There were {} sentences found. Here's a sample:".format(len(sentences)))
```

```python
# For each token, print corresponding part of speech tag
#for token in doc:
for token in doc:
    print('{} - {}'.format(token, token.pos_))
```

```python
for ent in doc.ents:
    print('{} - {}'.format(ent, ent.label_))
```

```python
# For every token in doc, print log-probability of the word, estimated from counts from a large corpus 
for token in doc:
    print(token, ',', token.prob)
```

```python
article = []
for w in doc:
    # if it's not a stop word or punctuation mark, add it to our article!
    if w.text != '\n' and not w.is_stop and not w.is_punct and not w.like_num:
        # we add the lematized version of the word
        article.append(w.lemma_)
print(article)
```

```python
keywords = Counter()
for chunk in doc.noun_chunks:
        if nlp.vocab[chunk.lemma_].prob < - 8:  # probability value -8 is arbitrarily selected threshold
            keywords[chunk.lemma_] += 1
print(keywords.most_common(5))            
```

```python
# all tokens that arent stop words or punctuations
words = [token.text for token in doc if token.is_stop != True and token.is_punct != True]

# noun tokens that arent stop words or punctuations
nouns = [token.text for token in doc if token.is_stop != True and token.is_punct != True and token.pos_ == "NOUN"]

# noun tokens that arent stop words or punctuations
ent = [ent for ent in doc.ents if ent.label_ == "PERSON" or ent.label_ == "GPE" or ent.label_ == "ORG" or ent.label_ == "LAW"]

# five most common tokens
ent_freq = Counter(ent)
common_ent = ent_freq.most_common(5)
# five most common tokens
word_freq = Counter(words)
common_words = word_freq.most_common(5)

# five most common noun tokens
noun_freq = Counter(nouns)
common_nouns = noun_freq.most_common(5)
```
